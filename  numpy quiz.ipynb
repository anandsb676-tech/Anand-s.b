```python
a=np.array([10,20,30,40,50])
mask=a>25
idx=np.array([2,3])
print("a[mask]:",a[mask])
print("idx:",a[idx])
```

```python
a=np.array([
    [1,2,3],
    [4,5,6]
])
b=np.array([[1,0,1],[0,1,0]],dtype=bool)
print(a[b])
```

```python
a=np.array([[1,2],[3,4],[5,6]])
x=a[a[:,1]%2==0]
print(a)
print(x)
```

```python
a=np.array([10,20,30,40,50])
idx=np.array([4,0,4,1])
x=a[idx]
print("A:",a)
print("x:",x)
```

```python
a=np.array([
    [1,2],
    [3,4],
    [5,6]
])
mask=np.array([True,False,True])
print(a[mask])
```

```python
a=np.array([[1,2,3],[4,5,6]])
b=np.array([10,20,30])
print(a+b)
```

```python
a=np.array([[1,2,3],[4,5,6]])
print((a>2)&(a<6))
x=a[(a>2)&(a<6)]
print(x)
```

```python
a=np.arange(50).reshape(10,5)
selected_rows=a[a[:,0]>20]
x=a[:,0]>20
print(selected_rows)
print(x)
```
