```python
#(np.sum(array),,,,)( np.mean(array) ,,,,,)(np.max(array),,)( np.min(array) ,,,np.argmax(b) >> b=np.array([10,40,30]) >>>>>index max value
import numpy as np


crop_yield=np.array([
    [12.5,13.2,14.1,13.8,15.2],# farm 1
    [10.3,11.5,10.8,12.1,11.9],#farm 2
    [15.1,14.8,16.2,15.5,16.8],#farm 3
    [11.2,10.9,12.3,11.7,12.8]#farm 4
  ])
print("CROP YIELDS ARRAY (TONS)")
print("-----------------------------------------------")
print(crop_yield)
print("-----------------------------------------------")
print(f"\nShape:{crop_yield.shape}")
print("-----------------------------------------------")
print(f"Data type:{crop_yield.dtype}")
print("-----------------------------------------------")
print(f"Total elements:{crop_yield.size}")
print("-----------------------------------------------")
print(f"Number of dimension:{crop_yield.ndim}")
print("-----------------------------------------------")
print("AGRICULTURAL ANALYSIS OPERATION")
print("-----------------------------------------------")
total_yield=np.sum(crop_yield)
avg_yield_per_farm=np.mean(crop_yield,axis=1)
avg_yield_per_season=np.mean(crop_yield,axis=0)
max_yield=np.max(crop_yield)
min_yield=np.min(crop_yield)
print(f"TOTAL YIELD:{total_yield}TONS")
print(f"AVERAGE PER FARM:{avg_yield_per_farm}TONS")
print(f"AVERAGE PER SEASON:{avg_yield_per_season}TONS")
print(f"MAXIMUN YIELD:{max_yield} tons")
print(f"MINIMUM YIELD:{min_yield}TONS")

best_farm=np.argmax(avg_yield_per_farm)
print(f"\nBEST PERFORMING FARM:{best_farm}")
print(f"Their average yield:{avg_yield_per_farm[best_farm]}")
```
