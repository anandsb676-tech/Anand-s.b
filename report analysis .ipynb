```python
#agriculture data analyis project
#------functiom defintions----

def load_data(filename):
  #"""reads crops data from a file and returns a list of dictionaries"""
  crops=[]
  files=open(filename,"r")
  for line in files:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "rainfall":float(data[3]),
          "fertilzer":float(data[4]),
          "yield":float(data[5])
      }
      crops.append(crop)
      files.close()
      return crop



def calculate_yield_per_area(crop):
  #"""returns yield per unit area"""
  return crop["yield"]/crop["area"]


def display_crops(crops):
  #displays all crop records
  print("\n crop records :")
  for crop in crops:
    ypa=calculate_yield_per_area(crop)
    print(crop["name"],"-",crop["season"],"-","yield/area:",round(ypa,2))

def find_highest_and_lowest_yield(crops):
  if not crops:
    return None, None # Handle case of empty crops list

  highest=crops[0]
  lowest=crops[0]

  for crop in crops:
    if calculate_yield_per_area(crop)>calculate_yield_per_area(highest):
      highest=crop
    if calculate_yield_per_area(crop)<calculate_yield_per_area(lowest):
      lowest=crop
  return highest,lowest

def seasonal_summary(crops):
  #"""generates sesonal summary yield using dictionary"""
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_area(crop)

    if season not in summary:
      summary[season]=[]
    summary[season].append(ypa)
  return summary


def write_report(filename,crops,highest,lowest,summary):
  with open(filename,"w") as file:
    files.write("agriculture data analysis report\n\n")

    files.write("crop yield per unit area :\n")
    for crop in crops:
      ypa=calculate_yield_per_area(crop)
      files.write(f"{crop['name']}: {round(ypa,2)}\n")
    files.write("\n") # Add a newline for separation

    if highest:
      files.write("Highest yield crop:\n")
      files.write(f"{highest['name']} (Yield/Area: {round(calculate_yield_per_area(highest),2)})\n")
    else:
      files.write("Highest yield crop: N/A\n")
    files.write("\n")

    if lowest:
      files.write("Lowest yield crop:\n")
      files.write(f"{lowest['name']} (Yield/Area: {round(calculate_yield_per_area(lowest),2)})\n")
    else:
      files.write("Lowest yield crop: N/A\n")
    files.write("\n")

    files.write("Seasonal average yield:\n")
    for season, ypa_list in summary.items():
      if ypa_list: # Ensure list is not empty before calculating average
        avg=sum(ypa_list)/len(ypa_list)
        files.write(f"{season}: {round(avg,2)}\n")
      else:
        files.write(f"{season}: No data\n")
```

```python
#main program
def main():
  input_file="crops.txt"
  output_file="analysis_report.txt"

  # Assuming load_data will be fixed to return a list of dictionaries
  crops = load_data(input_file)

  # Add a check for valid data, as load_data might still be incorrect
  if not isinstance(crops, list) or not crops:
    print("Error: Could not load crop data or data is empty. Please check 'load_data' function.")
    return

  # Display all crop records once
  display_crops(crops)

  # Find highest and lowest yield once
  highest, lowest = find_highest_and_lowest_yield(crops)

  # Generate seasonal summary once
  summary = seasonal_summary(crops)

  # Write the analysis report once
  write_report(output_file, crops, highest, lowest, summary)
  print("\n Analysis complete. Report written to ", output_file)

main()
```
