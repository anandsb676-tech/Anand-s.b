```python
import pandas as pd
data={
    "calories":[420,380,390],
    "duration":[50,40,45],


}
df=pd.DataFrame(data)
print(df)
print("\n")
print(df.loc[0])
print("\n")
print(df.loc[[0,1]])

```

```python
data={
    "calories":[420,380,390],
    "duration":[50,40,45]
}
df=pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

```python
data={
    "name":["alice","bob","charles"],
    "age":[20,21,22],
    "marks":[89,78,90]
}
df=pd.DataFrame(data)
print(df)
```
