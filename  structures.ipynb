```python
#===CONTROL STRUCTURES:IF-ELSE===
#1.Simple if statement
print("===simple IF===")
soil_moisture=float(input("enter soil moisture%;"))
if soil_moisture<30:
  print("soil is dry ! Irrigation needed.")
```

```python
#2.if-else
print("\n===IF-ELSE===")
temperature=float(input("enter temperature"))
if temperature>50:
  print("High temperature alert!")
  print("consider shade net for sensitive crops:")
else:
  print("normal temperature")
```

```python
#.if-elif-else
print("\n===IF-ELIF-ELSE===")
soil_ph = 6 # Example value, you might want to get this from input
if soil_ph < 5.5:
  print("soil is too acidic")
  print("recommendation: add lime")
elif soil_ph > 7.5:
  print("soil is too alkaline")
  print("recommendation :add sulphur")
else:
  print("soil ph is in optimal range (5.5-7.5)")
```

```python
#CROP RECOMMENDATION SYSTEM
print("\n===crop recommendation system===")
rainfall=float(input("enter rainfall in mm:"))
soil_type="loamy";
if rainfall  > 200 and soil_type == "clay":
  crop="rice"
elif rainfall > 100 and soil_type in["loamy","sandy","loamy"]:
  crop="wheat"
elif rainfall < 80:
  crop="millet(drought resistant)"
else:
  crop="mixed vegetables"
print("recommended crop:",crop)
```

```python
#nested if
print("\n===irrrigation decision system===")
moisture=int(input("enter moisture="))
rain_forecast=True
last_irrigation_days=int(input("enter last irrigated days:"))
if moisture < 40:
  if rain_forecast:
    print("moisture low,but rain expected")
    print("action:wait for rain")
  else:
   if last_irrigation_days > 2:
    print("irrigate now!")
   else:print("check tommorow")
else:
  print("moisture adequate,no irrigation")


```

```python
#Grade crop quality
print("\n===crop quality grading===")
weight = int(input("enter weight in grams < less than 100gm:"))
blemishes = int(input("enter  blemishes value 0-5"))
color_score = int(input("enter  color score out of 10:"))
if weight >= 80 and blemishes==0 and color_score >= 8:
  grade = "A (premiun)"
elif weight >= 60 and blemishes <= 2 and color_score >=6:
  grade = "B (standard)"
elif weight >=40:
  grade = ("c(processing)")
else:
  grade = "reject"
print(f"crop grade:{grade}")

```
