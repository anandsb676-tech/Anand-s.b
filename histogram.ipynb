```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D","E"])
y=np.array([3,8,25,10,7])
colors=["chartreuse","yellow","orange","khaki","aqua"]
plt.bar(x,y,color=colors)
plt.show()
```

```python
x=np.random.normal(170,1,100)

plt.hist(x)
plt.show()
```

```python
data=np.random.normal(170,10,250)
plt.hist(data,bins=10,color="green",edgecolor="black")
plt.xlabel("VALUE")
plt.ylabel("FREQUENCY")
plt.title("HISTOGRAM EXAMPLE")
plt.show()
```

```python
#generate random data
data=[2,4,5,7,8,10,12,13,15,17,18,20,22,24,25,27,29,30,32,35]
plt.hist(data,bins=5,edgecolor="black")
plt.title("Histogram with 5 Bins")
plt.show()
```

```python
data=[
    30,35,40,26,31,59,86,54,12,35,48,75,69,12,32,54,65,89,85,25,45,35,51,35,56
    ]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="red",edgecolor="black",alpha=0.5,density=True,range=(30,80)

)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram without random data")
plt.grid(True)
plt.tight_layout
```
