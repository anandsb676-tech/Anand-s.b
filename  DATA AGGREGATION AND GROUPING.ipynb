```python
import pandas as pd
data={
    'department':['HR','SALES','IT','SALES','IT','HR','SALES'],
    'employee':['A','B','C','D','E','F','G'],
    'salary':[50000,60000,45000,48000,70000,72000,71000],
    'experience':[2,3,4,5,6,7,5]

}
df=pd.DataFrame(data)
print(df)
```

```python

print(df.groupby(['department','experience']).sum())
```

```python
print(df.groupby('experience').sum())
```

```python
print(df.groupby('department')['salary'].sum())
```

```python
print(df.groupby(['department','experience']).sum())
```

```python
print(df.groupby('department').sum())
```

```python
print(df.groupby('department')['salary'].agg(['sum','mean','std']))
```

```python
print(df.groupby('department').agg({'salary':'sum','experience':'mean'}))
```

```python
df['dept_avg_salary']=df.groupby('department')['salary'].transform('mean')
print(df)
```

```python
print(df.groupby('department').filter(lambda X:len(X)>2))
```
