```python

#functions for addition
def add(x,y):
  return x+y
#functions for subration
def subtract(x,y):
  return x-y
#function for multiplication
def multiply(x,Y):
  return x*y
#functions for devision
def divide(x,y):
  if y==0:
    return "error!division by 0"
  else:
    return x/y
#functin to take input values from the user
def get_input():
  try:
    #take user input for first number
    x=float(input("enter the first number :"))
    #take seond number
    y=float(input("enter the second number :"))
    return x,y
  except ValueError:
    print("invalid input! please enter a valid  number ")
    return get_input()
#function to print the menu of operations
def print_menu():
  print("\nselect operation")
  print("1.add")
  print("2.subtract")
  print("3.multiply")
  print("4.divide")
  print("5.exit")
#main calculation functions
def calculation():
  while True:
    print_menu()
    choice=input("enter choice (1/2/3/4/5):")
    if choice=='5':
      print("exiting the calculator")
      break
#call input_function
    x,y=get_input()
#perform operation based on user choice
    if choice=='1':
     print(f"{x}+{y}={add(x,y)}")
    elif choice=='2':
      print(f"{x}-{y}={subtract(x,y)}")
    elif  choice=='3':
      print(f"{x}*{y}={multiply(x,y)}")
    elif choice=='4':
      print(f"{x}/{y}={divide(x,y)}")
    else:
      print("invalid input! please select a vadil operation ")
calculation()
print("done with program")



```

```python
def average_yield(yields):
  total=sum(yields)
  return total/len(yields)
crop_yields=[1200,1350,1100,1500,1234]
avg=average_yield(crop_yields)
print("average yield:",avg)
```

```python
def increase_yield(yields, increment):
  for i in range(len(yields)):
    yields[i]+=increment
crop_yields=[1200,1350,1100,1500]
increase_yield(crop_yields,100)
print(crop_yields)
```

```python
def convert_to_quintals(yields):
  result=[]
  for value in yields:
    result.append(value/100)
  return result
crop_yields=[1000,2341,1100,1560]
quintals=convert_to_quintals(crop_yields)
print(quintals)
```

```python
def high_yield_crop(yields,threshold):
  high=[]
  for value in yields:
    if value>=threshold:
     high.append(value)
  return high

crop_yields=[1200,500,1500,800]
print(high_yield_crop(crop_yields,100))



```

```python
def collect_yields():
  yields=[]
  n=int(input("enter number of fields:"))
  for i in range(n):
    value=int(input("enter yield:"))
    yields.append(value)
  return yields
data=collect_yields()
print("collected yields:",data)
```
