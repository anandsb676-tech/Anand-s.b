```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

data={
    'weather':['sunny','sunny','rainy','rainy','cloudy','sunny','cloudy','rainy'],
    'soil':['dry','wet','dry','dry','dry','wet','wet','dry'],
    'temperature':['hot','hot','hot','cold','hot','cold','cold','hot'],
    'irrigation':['yes','no','yes','no','yes','no','yes','no']
}

df=pd.DataFrame(data)
print("dataset:\n",df)
```

```python
le_weather=LabelEncoder()
le_soil=LabelEncoder()
le_temp=LabelEncoder()
le_target=LabelEncoder()

df['weather']=le_weather.fit_transform(df['weather'])
df['soil']=le_soil.fit_transform(df['soil'])
df['temp']=le_temp.fit_transform(df['temperature'])
df['irrigate']=le_target.fit_transform(df['irrigation'])

print("\nEncode Dataset:\n",df)
```

```python
X=df[["weather","soil","temp"]]
y=df["irrigate"]

model=CategoricalNB()
model.fit(X,y)
```

```python
test_data=pd.DataFrame({
    'weather':['rainy'],
    'soil':['dry'],
    'temp':['hot']
})

test_data['weather']=le_weather.transform(test_data['weather'])
test_data['soil']=le_soil.transform(test_data['soil'])
test_data['temp']=le_temp.transform(test_data['temp'])

prediction=model.predict(test_data)


result=le_target.inverse_transform(prediction)
print("prediction:",result)

prob=model.predict_proba(test_data)
print("\nProbability:\n",prob)




```
