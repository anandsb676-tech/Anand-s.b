```python
import random

def number_guessing_game():
    """
    A simple number guessing game.
    The computer picks a number and the user tries to guess it.
    """
    print("Hi! Welcome to the Number Guessing Game.")

    # Define the range for the random number
    low = 1
    high = 100

    # Generate a random number between the lower and upper bounds
    secret_number = random.randint(low, high)

    attempts = 0
    guess = None

    print(f"\nYou have to guess a number between {low} and {high}.")

    # The main game loop
    while guess != secret_number:
        try:
            # Get user input for their guess
            guess = int(input('Enter your guess: '))
            attempts += 1

            # Provide feedback on the guess
            if guess == secret_number:
                print(f'\nCorrect! The number is {secret_number}.')
                print(f'You guessed it in {attempts} attempts.')
                break # Exit the loop if the guess is correct
            elif guess < secret_number:
                print('Too low! Try a higher number.')
            else:
                print('Too high! Try a lower number.')
        except ValueError:
            # Handle cases where the user enters something that isn't an integer
            print("Invalid input. Please enter a whole number.")

if __name__ == "__main__":
    number_guessing_game()

```

```python
import random

def math_game():
    score = 0
    total_questions = 5
    operators = ['+', '-', '*']

    print("--- Welcome to the Mental Math Game! ---")
    print(f"Answer {total_questions} questions to get your score.\n")

    for i in range(1, total_questions + 1):
        # Generate two random numbers between 1 and 12
        num1 = random.randint(1, 12)
        num2 = random.randint(1, 12)
        operator = random.choice(operators)

        # Calculate correct answer
        if operator == '+':
            answer = num1 + num2
        elif operator == '-':
            answer = num1 - num2
        else: # Multiplication
            answer = num1 * num2

        # Ask the user
        try:
            user_input = int(input(f"Question {i}: {num1} {operator} {num2} = "))

            if user_input == answer:
                print("Correct! ✅")
                score += 1
            else:
                print(f"Wrong. The answer was {answer}. ❌")
        except ValueError:
            print("Invalid input, that counts as wrong! ❌")

    print(f"\nGame Over! Your final score is: {score}/{total_questions}")

# Run the game
math_game()

```

```python
import random

def play_21():
    current = 0
    while current < 21:
        player = int(input("Take 1-3 numbers: "))
        current += player
        if current >= 21: print("You lose!"); break
        comp =random.randint(1,3)# Basic strategy to win
        current += comp
        print(f"Computer takes {comp}. Total: {current}")
        if current >= 21: print("Computer loses! You win!"); break

if __name__ == "__main__": play_21()
```

```python
import random

def play_cricket():
    print("--- Welcome to Python Cricket! ---")
    print("Rules: Enter a number from 1 to 6. If it matches the computer's ball, you're OUT!")

    score = 0
    while True:
        try:
            player_move = int(input("Enter your shot (1-6): "))

            if player_move < 1 or player_move > 6:
                print("Invalid input! Please enter a number between 1 and 6.")
                continue

            computer_ball = random.randint(1, 6)
            print(f"Computer bowled: {computer_ball}")

            if player_move == computer_ball:
                print(f"--- HOWZAT! You're OUT! ---")
                break
            else:
                score += player_move
                print(f"Nice shot! Current Score: {score}")
                print("-" * 20)

        except ValueError:
            print("Please enter a valid number.")

    print(f"\nGAME OVER. Your final score is: {score}")

if __name__ == "__main__":
    play_cricket()
```

```python
import random

def play_innings(role, target=None):
    score = 0
    is_out = False

    print(f"\n--- Start of {role} Innings ---")
    if target:
        print(f"Target to beat: {target}")

    while not is_out:
        try:
            player = int(input(f"Your move (1-6): "))
            if not 1 <= player <= 6:
                print("Pick a number between 1 and 6!")
                continue

            comp = random.randint(1, 6)
            print(f"Computer: {comp}")

            if player == comp:
                print("--- OUT! ---")
                is_out = True
            else:
                # Add runs based on who is batting
                runs = player if role == "Batting" else comp
                score += runs
                print(f"Current Score: {score}")

                # If computer is chasing, check if they won
                if target and score >= target:
                    break
        except ValueError:
            print("Invalid input! Use numbers only.")

    return score

# --- Game Flow ---
# 1st Innings: You Bat
your_score = play_innings("Batting")
target = your_score + 1
print(f"\nYour final score: {your_score}")
print(f"The computer needs {target} to win.")

# 2nd Innings: Computer Bats
comp_score = play_innings("Bowling", target)

# Result
print("\n--- FINAL RESULT ---")
if comp_score >= target:
    print(f"Computer wins by {6 - 1} wickets! (Final Score: {comp_score})")
else:
    print(f"You win! You defended your score by {target - comp_score - 1} runs.")
```
