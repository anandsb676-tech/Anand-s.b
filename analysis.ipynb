```python
def load_data(filename):
  crops=[]
  files=open(filename,"r")

  for line in files:
   line.strip()
   if line:
    data=line.split(",")
    crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "revenue":float(data[5])
     }
    crops.append(crop)
  return crops

def find_all_kharif_crops(crops):
  kharif=[]
  for crop in crops:
    if crop["season"]=="kharif":
      kharif.append(crop)

  return kharif

def find_all_non_kharif_crops(crops):
  non_kharif=[]
  for crop in crops:
    if crop["season"]!="kharif":
      non_kharif.append(crop)

  return non_kharif

def calculate_profit(crops):
  for crop in crops:
    print("\n",crop["name"],"crop profit")
    print(crop["revenue"]-crop["cost"])

  return


def total_area(crops):
  for crop in crops:
   total_area=crop["area"]
  return total_area











```

```python
def main():
  input_file="abcd (1).txt"
  crops=load_data(input_file)
  print(crops)
  kharif_crop=find_all_kharif_crops(crops)
  print(kharif_crop)
  non_kharif_crop=find_all_non_kharif_crops(crops)
  print(non_kharif_crop)
  profit=calculate_profit(crops)
  print(profit)
  all_area=total_area(crops)
  print(all_area)
main()
```
