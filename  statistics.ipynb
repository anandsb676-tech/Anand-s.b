```python
import numpy as np
#rainfall data(mm) over 12 months
rainfall=np.array([45,52,68,120,180,220,195,140,110,75,58,40])
#corresponding crop yield(tons/hectare)
yields=np.array([2.1,2.3,2.8,4.2,5.8,6.5,5.9,5.2,3.8,2.9,2.4,2.0])
print("RAINFALL")
print(rainfall)
print("YIELDS")
print(yields)
```

```python
print("===RAINFALL STATISTICS===")
print(f"mean:{np.mean(rainfall):.2f}mm")
print(f"median:{np.median(rainfall):.2f}mm")
print(f"standard deviation:{np.std(rainfall):.2f}")
print(f"variance:{np.var(rainfall):.2f}")
print(f"min:{np.min(rainfall)}mm (month{np.argmin(rainfall)+1})")
print(f"max:{np.max(rainfall)}mm (month{np.argmax(rainfall)+1}")
print(f"Range:{np.ptp(rainfall)}mm")
```

```python
print("===YIELD STATISTICS===")
print(f"mean:{np.mean(yields):.2f}")
print(f"median:{np.median(yields):.2f}")
print(f"std dev:{np.std(yields):.2f}")
```

```python
print("===PERCENTILES===")
percentiles=[25,50,75,90]
rainfall_percentiles=np.percentile(rainfall,percentiles)
for p,val in zip(percentiles,rainfall_percentiles):
 print(f"{p}th percentile rainfall:{val:.2f}mm")
```

```python
print("===CORRELATION ANALYSIS===")
correlation_matrix=np.corrcoef(rainfall,yields)
print('\ncorrelation_matrix==')
print(correlation_matrix)
correlation=correlation_matrix[0,1]
print('\ncorrelation==')
print(correlation)
print(f"\nrainfall-yields corelation:{correlation:.4f}")
print("interpretation:strong positive correlation"if correlation>0.7 else "moderate coorelation"  if correlation>0.4 else"eak correlation")
```

```python
print("===CUMULATIVE RAINFALL===")
cumulative=np.cumsum(rainfall)
print("month\trainfall\tcumulative")
for month,(rain,cum) in enumerate(zip(rainfall,cumulative),1):
  print(f"{month}\t{rain}\t\t{cum}")
```
