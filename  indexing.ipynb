```python
#index method
#returns the index of the first occurence of an element
thislist=["apple","banana","cherry","banana"]
index_value=thislist.index("banana")
print("index of banana:",index_value)
```

```python
#count()method
#count how many times an element appears in the list
thislist=["apple","banana","apple","cherry","apple"]
count_value=thislist.count("apple")
print("count of apple :",count_value)
```

```python
#reverse () method
# reverse the order of elements in the list (in-place)
thislist=["orange","mango","kiwi","banana","apple"]
print("original list:",thislist)
thislist.reverse()
print("reverse list:",thislist)
```

```python
# daily farm expenses
expenses=[
    ["day1",[1200,500,300]],
    ["day2",[1500,400,600]],
    ["day3",[1000,700,200]]
]
print(expenses)
print(expenses[0])
for day in expenses:
  print(day[0],"total:",sum(day[1]))
```

```python
def expenses_made():
  expenses=[]
  number_of_days=int(input("enter the number of days :"))
  for i in range(number_of_days):
    day_name=(input(f"enter day name:"))
    seed_cost=int(input("enter seed cost:"))
    fertilizer_cost=int(input("enter fertilzer cost:"))
    labour_cost=int(input("enter labour cost:"))
    expenses.append([day_name,[seed_cost,fertilizer_cost,labour_cost]])
  return expenses
total_expenditure=expenses_made()
print("daily expenses")
for expenditure in total_expenditure:
  print(expenditure[0],"expenses are",expenditure[1])
```
