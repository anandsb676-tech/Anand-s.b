```python
import pandas as pd
data={
    'farm_id':['F001','F002','F003','F004','F005','F006','F007'],
    'crop':['Wheat','Rice','Corn','Wheat','Rice','Corn','Wheat'],
    'yield_tons':[180,228,416,164,245,390,195],
    'rainfall_mm':[450,820,680,420,850,710,480],
    'fertilizer_kg':[120,150,200,110,160,195,125],
    'soil_ph':[6.5,5.8,6.2,6.7,5.9,6.1,6.4]
}

df=pd.DataFrame(data)
df['yield_per_ha']=(df['yield_tons']/50).round(1) #assume 50 ha each
print("ORIGINAL DATASET")
print(df)

#filter1:high yield farms
print("\n===HIGH YIELD FARMS(>200 TONS)===")
high_yield=df[df['yield_tons']>200]
print(high_yield)


#filter2:multiple conditions(rice with high rainfall))
print("\n ===RICE WITH HIGH RAINFALL(>800MM)===")
rice_high_rain=df[(df['crop']=='Rice')&(df['rainfall_mm']>800)]
print(rice_high_rain)

#filter3 using isim for scepicific crops
print("\n===GRAIN CROPS(WHEAT&RICE0===")
grain_crops=df[df['crop'].isin(['Wheat','Rice'])]
print(grain_crops[['farm_id','crop','yield_tons']])


#filet4 range filtering
print("\n===OPTIMAL SOIL PH(6.0,6.5)")
optimal_ph=df[df['soil_ph'].between(6.0,6.5)]
print(optimal_ph[['farm_id','soil_ph','yield_tons']])






#filtering query method
print("\n===COMPLEX QUERY:GOOD CONDITIONS===")
min_rainfall=600
good_conditions=df.query('rainfall_mm>@min_rainfall and fertilizer_kg>150')
print(good_conditions[['farm_id','crop','rainfall_mm','fertilizer_kg']])


#position based selections
print('\n===FIRST 3 FARMS(iloc)===')


#label based selections
print("\n===SPECIFIC COLUMNS(loc)===")
print(df.loc[:,['farm_id','crop','yield_per_ha']])

#combined filyering and selections
print("\n===TOP PERFORMERS - SELECTED COLUMNS===")
top=df[df['yield_per_ha']>4.5][['farm_id','crop','yield_per_ha']]
print(top)

```
