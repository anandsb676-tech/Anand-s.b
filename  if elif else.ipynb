```python
# crop siutability check
#accept soil moisture(%) and raingall(MMM) from user
print("crop suitability check")
moisture=int(input("enter soil moisture in % :"))
rainfall=int(input("enter rainfall in mm :"))
if moisture>=70 and rainfall>=200:
 print("suitable crop is rice")
else: print("not suitable for rice")
```

```python
#crop yield classification
print("crop yield classification")
crop_yield=float(input("enter crop yield in quintals : "))
if crop_yield>=50:
  print("high yield")
elif crop_yield<30:
  print("low yield")
else: print("medium yield ")
```

```python
#fertilizer recommendation system
print(" fertilizer recommendation system ")
crop_name=input("enter crop name:").lower()
soil_nitrogen_level=float(input("enter soil nitrogen level :"))
if crop_name=="rice" and soil_nitrogen_level < 50:
  print("apply urea!")
elif crop_name=="wheat" and soil_nitrogen_level < 40:
  print("apply DAP :")
else: print("fertilser not required !!!!")
```

```python
#crop availability check
crops=["rice","wheat","maize"]
crop_name=input("enter crop name:").lower()
if crop_name in crops:
    print("===crop is avaiable===")
else: print("===not available===")
```

```python
#average rainfall analysis
print("===rainfall analysis===")
first_day=float(input("enter day 1 rainfall="))
second_day=float(input("enter  day 2 rainfall="))
third_day=float(input("enter day 3 rainfall="))
fourth_day=float(input("enter day 4 rainfall="))
fifth_day=float(input("enter day 5 rainfall="))
average_rainfall=first_day+second_day+third_day+fourth_day+fifth_day/5
print("average rainfall= ",average_rainfall)
if average_rainfall>=100:
  print("suficient rainfall!!!!!")
else: print("insufficient rainfall")
```

```python
#seed quality evaluation
print("===SEED QUALITY EVALUATION===")
gp=float(input("enter germination % :"))
mois=float(input("enter seed moisture % :"))
if gp>=85:
  if mois<=12:
    print("high quality seed")
  else: print("moisture too high !!!!")
else: print("low germination % !!!")
```
