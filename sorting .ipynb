```python

import numpy as np
import pandas as pd


df=pd.DataFrame({
    'NAME':["ALICE","BOB","CHARLIE","DAVID","EVA","FRANK"],
    "DEPARTMENT":["HR","IT","IT","FINANCE","HR","FINANCE"],
    "SALARY":[50000,30000,70000,60000,50000,60000],
    "EXPERIENCE":[2,5,4,6,3,6],

})
print(df)
```

```python
sal=df.sort_values('SALARY')
print(sal)
```

```python
sal=df.sort_values(['SALARY','EXPERIENCE'],ascending=[False,True])
print(sal)
```

```python
des=df.sort_index(ascending=False)
print(des)
```

```python
df['rank_salary']=df['SALARY'].rank()
print(df)
```

```python
df['rank_salary']=df['SALARY'].rank(ascending=False)
print(df)
```

```python
df['dense_rank_slary']=df['SALARY'].rank(ascending=True,method='dense')
print(df)
```

```python
da=df['DEPARTMENT'].value_counts()
print(da)
```

```python
df=df['DEPARTMENT'].value_counts(normalize=True)
print(df)
```

```python
import matplotlib.pyplot as plt


df=pd.DataFrame({
    'NAME':["ALICE","BOB","CHARLIE","DAVID","EVA","FRANK"],
    "DEPARTMENT":["HR","IT","IT","FINANCE","HR","FINANCE"],
    "SALARY":[50000,30000,70000,60000,50000,60000],
    "EXPERIENCE":[2,5,4,6,3,6],

})


df[['SALARY','EXPERIENCE']].plot()
plt.show()
```

```python
df["SALARY"].plot(kind='hist',bins=5)
plt.show()
```
