```python
import pandas as pd
s=pd.Series([10,20,30,40,80])
for val in s:
  print(val)

for idx,val in s.items():
  print(idx,val)





```

```python
import pandas as pd
s=pd.Series([100,200,300],index=['a','b','c'])
print(s)

for label,value in s.items():
  print(label,value)

for label in s.index:
  print(label)

for value in s.values:
  print(value)
```
